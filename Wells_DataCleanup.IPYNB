```python
import geopandas as gpd
import pandas as pd
import fiona
```

```python
path_gpkg= "../GEOPKG/Thematic_Data.gpkg"
gdf = gpd.read_file(path_gpkg, layer='well_agg_centroid_OZ_')
gdf.head()
```

```python
# Example groupby operation
grouped = gpd.GeoDataFrame.groupby(gdf, "Name").agg({
    'XCOORD': 'mean',   # Replace 'attribute1' with the actual column name
    'YCOORD': 'mean',   # Replace 'attribute2' with the actual column name
    'Num_Wells': 'sum',   # Replace 'attribute3' with the actual column name
    'Q_P50_sum' : 'sum',
    'Q_AVERAGE_sum' : 'sum',
    'TOP_fltr_max' : 'max',
    'BOTTOM_fltr_min' : 'min',
    'geometry' : lambda x: x.unary_union.centroid
}).reset_index()
```

```python
grouped =gpd.GeoDataFrame(
    grouped, geometry=grouped.geometry, crs="EPSG:28992")

grouped
```

```python
wellInfo = gpd.read_file('../CSV/Vitens_watersupply_Overijssel.csv')
wells = wellInfo.drop(columns="geometry")
wells.head()
```

```python
merged = gpd.GeoDataFrame.merge(grouped,wells, left_on="Name", right_on="Location")
```

```python
merged['Q_median_YR'] = (merged.Q_P50_sum*-365/1000000)
merged['Q_AVG_YR']  = (merged.Q_AVERAGE_sum*-365/1000000)
merged = merged.rename(columns={
    'Q_P50_sum' : 'Q_Median_d',
    'Q_AVERAGE_sum': 'Q_AVG_d'
    })
merged.head()
```

```python
merged.to_file(path_gpkg, layer='Well_Capacity', driver="GPKG")
```
